```python
import pandas as pd
import numpy as np
```

###  数据重采样 ###
- 时间数据由一个频率转换到另一个频率
- 降采样
- 升采样

```python
rng = pd.date_range('1/1/2011', periods=90, freq='D') #随机生成90天的数据
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts.head()
```

```python
ts.resample('M').sum() #以月为单位进行采样后求和
```

```python
ts.resample('3D').sum().head(8) #以三天为单位进行采样后求和
```

```python
day3Ts = ts.resample('3D').mean()
day3Ts
```

```python
#day3Ts数据是以三天为单位的，现在想变成以天为单位的数据
#进行升采样（差值填充）
print(day3Ts.resample('D').asfreq())
```

###  插值方法： ###
- ffill 空值取前面的值
- bfill 空值取后面的值
- interpolate 线性取值

```python
day3Ts.resample('D').ffill(1)
```

```python
day3Ts.resample('D').bfill(1)
```

```python
day3Ts.resample('D').interpolate('linear')
```
